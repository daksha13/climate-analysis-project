```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# load data
data = pd.read_csv("/data/climate_cleaned.csv")

# preview the data and make sure everything loaded correctly
print(data.head())
print(data.info())
```

```python
# create a line plot to show average temperatures change over the years 
plt.figure(figsize=(12,6))
sns.lineplot(x='YEAR', y='ANNUAL_MEAN', data=data, marker='o')

# add titles and labels
plt.title('India Annual Mean Temperature (1901-2021)')
plt.xlabel('Year')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.tight_layout()
plt.savefig('India Annual Mean Temperature (1901-2021)')
plt.show()
```

```python
# create a new column to represent the decade each year falls in
data['Decade'] = (data['YEAR'] //10) * 10

# group by decade and calculate the average annual tempertature for each one
decade_avg = data.groupby('Decade')['ANNUAL_MEAN'].mean().reset_index()

# create a bar chart to show average temperature per decade
plt.figure(figsize=(12,6))
sns.barplot(x='Decade', y='ANNUAL_MEAN', data=decade_avg, palette='coolwarm')

# add titles and labels
plt.title('Decade-wise average annual temperature in India')
plt.xlabel('Decade')
plt.ylabel('Avg temperature (°C)')
plt.xticks(rotation=45) # rotate x-axis labels for readibility
plt.tight_layout()
plt.savefig('Decade-wise average annual temperature in India')
plt.show()
```
